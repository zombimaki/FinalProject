```python
import pandas as pd
```

```python
# Read in first file - data from Jan 2010 to June 2018
incidents_b42018 = pd.read_csv('yearly_incident_data/pre_2018/incidents_2010_2018June.csv')

incidents_b42018.head()
```

```python
# Select fields to keep
incidents_b42018_clean = incidents_b42018[['PublicAddress', 'CCN', 'Precinct', 'BeginDate', 'Offense', 'Description', 'UCRCode', 'Lat', 'Long', 'Neighborhood']]
incidents_b42018_clean.head()
```

```python
# Read in second file - data from June 2018 to end of 2020
incidents_18to20 = pd.read_csv('yearly_incident_data/since_2018/incidents_2018June_2020.csv')
incidents_18to20.head()
```

```python
# Select fields to keep
incidents_18to20_clean = incidents_18to20[['publicaddress', 'caseNumber', 'precinct', 'beginDate', 'beginTime', 'offense', 'description', 'UCRCode', 'centerLat', 'centerLong', 'neighborhood']]
incidents_18to20_clean.head()
```

```python
# Rename fields
incidents_18to20_clean.rename({'publicaddress': 'publicAddress', 'beginDate': 'date', 'beginTime': 'time', 'UCRCode': 'ucrCode', 'centerLat': 'lat', 'centerLong': 'lon'}, axis=1, inplace=True)
incidents_18to20_clean.head()
```

```python
# Rename fields
incidents_b42018_clean.rename({'PublicAddress': 'publicAddress', 'CCN': 'caseNumber', 'Precinct': 'precinct', 'BeginDate': 'date', 'Time': 'time', 'Offense': 'offense', 'Description': 'description', 'UCRCode': 'ucrCode', 'Lat': 'lat', 'Long': 'lon', 'Neighborhood': 'neighborhood'}, axis=1, inplace=True)
incidents_b42018_clean.head()
```

```python
# Clean up the caseNumber field to remove spaces and be same format as post-2018 data
incidents_b42018_clean['caseNumber'] = incidents_b42018_clean['caseNumber'].str.replace(' ','')
incidents_b42018_clean.head()
```

```python
# Clean up precinct field
precinct_data = incidents_b42018_clean['precinct'].to_list()
# print(precinct_data)

precinct_num = []

for precinct in precinct_data:
    precinct_str = str(precinct)
    num_splits = precinct_str.split('.')
    num_only = num_splits[0]
    precinct_num.append(num_only)

print(precinct_num)
```

```python
# Insert cleaned up precinct back into column
incidents_b42018_clean['precinct'] = precinct_num

incidents_b42018_clean['precinct'] = incidents_b42018_clean['precinct'].astype('str')
incidents_b42018_clean.head()
```

```python
# Change 'date' field from string to datetime then split into seperate columns
incidents_b42018_clean['date'] = pd.to_datetime(incidents_b42018_clean['date'])
incidents_b42018_clean['incidentDate'] = incidents_b42018_clean['date'].dt.strftime('%Y/%m/%d')
incidents_b42018_clean['incidentTime'] = incidents_b42018_clean['date'].dt.strftime('%H:%M')
incidents_b42018_final = incidents_b42018_clean.drop(['date'], axis=1)
incidents_b42018_final.head()
```

```python
incidents_b42018_final.dtypes
```

```python
# Clean up case in string fields
incidents_b42018_final['publicAddress'] = incidents_b42018_final['publicAddress'].str.lower()
incidents_b42018_final['publicAddress'] = incidents_b42018_final['publicAddress'].str.title()
incidents_b42018_final['description'] = incidents_b42018_final['description'].str.lower()
incidents_b42018_final['description'] = incidents_b42018_final['description'].str.title()
incidents_b42018_final['neighborhood'] = incidents_b42018_final['neighborhood'].str.lower()
incidents_b42018_final['neighborhood'] = incidents_b42018_final['neighborhood'].str.title()
incidents_b42018_final.head()
```

```python
incidents_b42018_final.dtypes
```

```python
incidents_18to20_clean.head()
```

```python
# Clean up date field
incidents_18to20_clean['date'] = pd.to_datetime(incidents_18to20_clean['date'])
incidents_18to20_clean['incidentDate'] = incidents_18to20_clean['date'].dt.strftime('%Y/%m/%d')
incidents_18to20_cleandate = incidents_18to20_clean.drop(['date'], axis=1)
incidents_18to20_cleandate.head(10)
```

```python
# Push times to list for cleanup
times_raw = incidents_18to20_cleandate['time'].tolist()
# print(times_raw)
```

```python
# Fill out field to four digits
times_fixed = []

for time in times_raw:
    time_str = str(time)
    time_fixed = time_str.zfill(4)
    times_fixed.append(time_fixed)

# print(times_fixed)
```

```python
# Insert a colon into the time field
time_final = []

for time in times_fixed:
    digits = list(time)
    digits.insert(2,':')
    time_joined = ''.join(digits)
    time_final.append(time_joined)

# print(time_final)
```

```python
# Insert correct time back into the dataframe and drop the old time
incidents_18to20_cleandate['incidentTime'] = time_final
incidents_18to20_final = incidents_18to20_cleandate.drop(['time'], axis=1)
incidents_18to20_final.head()
```

```python
# Clean up UCR Code field to be single digit
ucr_data = incidents_18to20_final['ucrCode'].to_list()

ucr_num = []

for ucr in ucr_data:
    ucr_str = str(ucr)
    num_splits = ucr_str.split('.')
    num_only = num_splits[0]
    ucr_num.append(num_only)

# print(ucr_num)
```

```python
# Insert corrected UCR code back into the df
incidents_18to20_final['ucrCode'] = ucr_num
incidents_18to20_final
```

```python
# Clean up precinct field
precinct_data = incidents_18to20_clean['precinct'].to_list()
# print(precinct_data)

precinct_num = []

for precinct in precinct_data:
    precinct_str = str(precinct)
    num_only = precinct_str.replace(' ', '')
    # num_only = num_splits[0]
    precinct_num.append(num_only)

print(precinct_num)
```

```python
# Insert corrected precinct num back into the df
incidents_18to20_final['precinct'] = precinct_num
incidents_18to20_final['precinct'] = incidents_18to20_final['precinct'].astype('str')
incidents_18to20_final.head()
```

```python
# Clean up case in string fields
incidents_18to20_final['publicAddress'] = incidents_18to20_final['publicAddress'].str.lower()
incidents_18to20_final['publicAddress'] = incidents_18to20_final['publicAddress'].str.title()
incidents_18to20_final['description'] = incidents_18to20_final['description'].str.lower()
incidents_18to20_final['description'] = incidents_18to20_final['description'].str.title()
incidents_18to20_final['neighborhood'] = incidents_18to20_final['neighborhood'].str.lower()
incidents_18to20_final['neighborhood'] = incidents_18to20_final['neighborhood'].str.title()
incidents_18to20_final.head()
```

```python
# Combine all incidents into one csv file
all_incidents = pd.concat([incidents_b42018_final, incidents_18to20_final], ignore_index=True, sort=False)
all_incidents
```
